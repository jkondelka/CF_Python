```python
# defining a function with if statement: example 

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df = df.drop(columns = ['_merge'])
```

```python
# calling a function 

df['price_range'] = df.apply(price_label, axis=1)
```

### 4 : When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled â€œMost orders,â€ â€œAverage orders,â€ and â€œFewest orders.â€ Create a new column containing these labels called â€œbusiest_period_of_day.â€

```python
# run describe on order_hour_of_day column to get the feel about the values 
ords_prods['order_hour_of_day'].describe()
```

min order hour is 0 and max is 23, that fits as it is within a span of a day

```python
# count values in order_hour_of_day
ords_prods['order_hour_of_day' ].value_counts (dropna = False)
```

write __for-loop function__ that will go through the column and assign 

__Most orders__ - hours where orders exceed 2,000,000
__Average orders__ - hours where orders are between 1,000,000 and 2,000,000
__Fewest orders__ - hours where orders are less than 1,000,000


```python
# write a for-loop that sorts out values in order_hour_of_day based on above criteria

result3 = []

for value in ords_prods["order_hour_of_day"]:
  if value >= 8 and value <= 17:
    result3.append("Most orders")
  elif value == 8 or value == 18 or value == 19:
    result3.append("Average orders")
  else:
    result3.append("Fewest orders")
```

```python
# check the result
result3
```

```python
# create a new column called Busiest_period_of_day and assign result3 values to it
ords_prods['Busiest_period_of_day'] = result3
```
