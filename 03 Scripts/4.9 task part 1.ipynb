### check for data mixed data types, missing values and duplicates and address it

```python
# mixed data types

for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

```python
# change first_name data type to STRING 

customers['first_name'] = customers['first_name'].astype('str')
```

```python
# find missing values 
customers.isnull().sum()
```

```python
# display duplicates by creating subset that shows only duplicates 

dups = customers[customers.duplicated()]
```
